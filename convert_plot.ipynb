```python
import xarray as xr
import csv
import pandas as pd
import numpy as np
import matplotlib as plt


wind_nc_file = '/Users/michelleyu/Desktop/FALL2022/CSCI4350/project/wind/2021_01_01.nc4'
ds = xr.open_dataset(wind_nc_file)
df = ds.to_dataframe()

df = df.query('lon >= -3.14 and lon <= 1.3 and lat >= 51.15  and lat <= 54.74')
df = df.groupby(level = ['lat', 'lon'])['PS'].mean()

for i in range(2,13):
    if i< 10:
        wind_nc_file = '/Users/michelleyu/Desktop/FALL2022/CSCI4350/project/wind/2021_0'+str(i)+'_01.nc4'
        now = '2021_0'+str(i)+'_01.nc4'
        
    else:
        wind_nc_file = '/Users/michelleyu/Desktop/FALL2022/CSCI4350/project/wind/2021_'+str(i)+'_01.nc4'
        now = '2021_'+str(i)+'_01.nc4'
        

    if i < 11:
        before = '2021_0'+str(i-1)+'_01.nc4'
        
    else:
        before = '2021_'+str(i-1)+'_01.nc4'

    ds1 = xr.open_dataset(wind_nc_file)
    df1 = ds1.to_dataframe()

    #df1 = df1.query('lon >= 34 and lon <= 36 and lat >= 36 and lat <= 40')
    df1 = df1.query('lon >= -3.14 and lon <= 1.3 and lat >= 51.15  and lat <= 54.74')

    df1 = df1.groupby(level = ['lat', 'lon'])['PS'].mean()

    df = pd.merge(df, df1, how='inner', on=['lat', 'lon'], suffixes=(before, now))

```
